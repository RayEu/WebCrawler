```python
import pandas as pd
import os

def findfile(keyword, root):
    filelist=[]
    for root,dirs,files in os.walk(root):
        for name in files:  
            path = os.path.join(root, name)#绝对路径
            if (os.path.isfile(path)) & (keyword in path):
                filelist.append(os.path.join(root, name))
    return filelist

path = r'C:\Users\acer\Desktop\comment_analysis'
path = os.getcwd()
os.chdir(path)#切换路径
filelist = findfile('综艺',path)
commentData = pd.DataFrame()
for file in filelist:
    data = pd.read_excel(file)
    commentData = pd.concat([commentData, data])#多表合并
commentData = commentData.drop(labels='Unnamed: 0', axis=1)#按列名（axis=1）值（labels='Unnamed: 0'）丢弃
```

```python
#生成词云
import matplotlib.pyplot as plt
from wordcloud import WordCloud
def generate_wordcloud(text):
    # 生成一个词云图像
    #“词云”就是对网络文本中出现频率较高的“关键词”予以视觉上的突出，
    #形成“关键词云层”或“关键词渲染”，从而过滤掉大量的文本信息，
    #使浏览网页者只要一眼扫过文本就可以领略文本的主旨
    wordcloud = WordCloud(
        #设置字体，不指定就会出现乱码
        font_path=" C:\\Windows\\Fonts\\STXINGKA.TTF",
        #设置背景色
        background_color='white',
        #词云形状
        #mask=color_mask,
        #允许最大词汇
        max_words=5000,
        #最大号字体
        max_font_size=40
    ).generate_from_frequencies(text)#根据词频来画图， https://blog.csdn.net/MIKASA3/article/details/80066001
    # { word1: fre1, word2: fre2,  word3: fre3,......,  wordn: fren }
    
    # matplotlib的方式展示生成的词云图像
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
def collect_specified(comments, name):
    df = pd.DataFrame()
    for n in name:
        c = comments[comments['内容'].str.contains(n)]
        df = pd.concat([df, c])
    df = df.drop_duplicates()
    return df
```

```python
def zhanbi(value, label):
    values = [1-sum(value),]
    values.extend(value)
    labels = ['其他',]
    labels.extend(label)
    x = range(1, len(values)+1)
    #画柱状图
    plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一（替换sans-serif字体）
    plt.rcParams['axes.unicode_minus'] = False
    fig, ax = plt.subplots()
    for i in range(len(x)):
        b = ax.bar(x=x[i], height=values[i], width=0.4,
                   label=labels[i], color='darkorange')
        #h = b[0].get_height()
        h = values[i]
        ax.text(b[0].get_x()+b[0].get_width()/2, h,
                '%.2f%%'%(h*100), ha='center', va='bottom')
    plt.xticks(x, labels)
    plt.show()
```

```python
#六、情感分析
"""
1、使用snowNLP进行情感分析
https://www.cnblogs.com/otakurice/p/7638017.html
速度快，但分词差，情感分析效果一般
pip install snowNLP
2、使用腾讯云进行情感分析
申请：https://cloud.tencent.com/document/api/213/30654
https://blog.csdn.net/qq_38984677/article/details/83754222
因为要调用url访问腾讯云，速度慢
"""
from QcloudApi.qcloudapi import QcloudApi
#解析一条弹幕的情感
def get_one_danmu_TextSentiment_from_TX(content):    
    # 设置需要加载的模块
    module = 'wenzhi'
    # 接口名-情感分析
    action = 'TextSentiment'
    # 云API的公共参数-参数排序很重要(首字母排序)
    config = {
        'method': 'GET',
        'Region': 'ap-guangzhou',
        'secretId': 'AKIDhtqWFnU7uSo8QWD5VinOkskNkMo8c3g2',
        'secretKey': 'HpBIDqEK94j8ON7vYwEH3YnlZFm5B1cy',
        'SignatureMethod': 'HmacSHA1'
       }
    params = {'content':content, 'type':4}
    service = QcloudApi(module, config) 
    # 生成请求的url,但不发起请求
    service.generateUrl(action, params)
    # 调用接口-发起请求
    s = service.call(action, params).decode('utf-8')
    s = eval(s)
    return s["positive"], s['negative']
```

```python
#分析一个df的弹幕的情感
def get_df_TextSentiment(df):
    d = pd.DataFrame()
    #用户名	内容	会员等级	弹幕时间点	弹幕点赞	弹幕id	集数
    for index, c in df.iterrows():
        name = c['用户名']
        content = c['内容']
        timepoint = c['弹幕时间点']
        upcount = c['弹幕点赞']
        comment_id = c['弹幕id']
        period = c['集数']
        '''
        p, n = get_one_danmu_TextSentiment(content)
        cache = pd.DataFrame({'用户名':[name],'内容':[content], '弹幕时间点':[timepoint],
                              '弹幕点赞':[upcount],'弹幕id':[comment_id],'集数':[period],
                              '积极':[p], '消极':[n]})
        '''
        v = get_one_danmu_TextSentiment_from_snowNLP(content)
        cache = pd.DataFrame({'用户名':[name],'内容':[content], '弹幕时间点':[timepoint],
                              '弹幕点赞':[upcount],'弹幕id':[comment_id],'集数':[period],
                              '态度':[v]})
        d = pd.concat([d,cache])
    return d
```

```python
res = get_df_TextSentiment(baiyu_danmu)
res
```

```python
#排序
def cumsum_seq(v):
    sub = v.sort_values('弹幕时间点')
    return sub.loc[:, ['用户名', '内容', '弹幕时间点', '弹幕点赞', '弹幕id', '态度']]

res = res.groupby(['集数']).apply(cumsum_seq).reset_index().drop(columns='level_1')
res
```

```python
import matplotlib.pyplot as plt
#画图展示
tmp = res.groupby(['集数']).mean().reset_index()
plt.plot(tmp['集数'], tmp['态度'], marker='.')
plt.ylim(0, 1)
plt.show()
```
