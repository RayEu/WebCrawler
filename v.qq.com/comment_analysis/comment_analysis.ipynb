```python
import pandas as pd
import os

def findfile(keyword, root):
    filelist=[]
    for root,dirs,files in os.walk(root):
        for name in files:  
            path = os.path.join(root, name)#ç»å¯¹è·¯å¾„
            if (os.path.isfile(path)) & (keyword in path):
                filelist.append(os.path.join(root, name))
    return filelist

path = r'C:\Users\acer\Desktop\comment_analysis'
path = os.getcwd()
os.chdir(path)#åˆ‡æ¢è·¯å¾„
filelist = findfile('ç»¼è‰º',path)
commentData = pd.DataFrame()
for file in filelist:
    data = pd.read_excel(file)
    commentData = pd.concat([commentData, data])#å¤šè¡¨åˆå¹¶
commentData = commentData.drop(labels='Unnamed: 0', axis=1)#æŒ‰åˆ—åï¼ˆaxis=1ï¼‰å€¼ï¼ˆlabels='Unnamed: 0'ï¼‰ä¸¢å¼ƒ
```

```python
#ç”Ÿæˆè¯äº‘
import matplotlib.pyplot as plt
from wordcloud import WordCloud
def generate_wordcloud(text):
    # ç”Ÿæˆä¸€ä¸ªè¯äº‘å›¾åƒ
    #â€œè¯äº‘â€å°±æ˜¯å¯¹ç½‘ç»œæ–‡æœ¬ä¸­å‡ºç°é¢‘ç‡è¾ƒé«˜çš„â€œå…³é”®è¯â€äºˆä»¥è§†è§‰ä¸Šçš„çªå‡ºï¼Œ
    #å½¢æˆâ€œå…³é”®è¯äº‘å±‚â€æˆ–â€œå…³é”®è¯æ¸²æŸ“â€ï¼Œä»è€Œè¿‡æ»¤æ‰å¤§é‡çš„æ–‡æœ¬ä¿¡æ¯ï¼Œ
    #ä½¿æµè§ˆç½‘é¡µè€…åªè¦ä¸€çœ¼æ‰«è¿‡æ–‡æœ¬å°±å¯ä»¥é¢†ç•¥æ–‡æœ¬çš„ä¸»æ—¨
    wordcloud = WordCloud(
        #è®¾ç½®å­—ä½“ï¼Œä¸æŒ‡å®šå°±ä¼šå‡ºç°ä¹±ç 
        font_path=" C:\\Windows\\Fonts\\STXINGKA.TTF",
        #è®¾ç½®èƒŒæ™¯è‰²
        background_color='white',
        #è¯äº‘å½¢çŠ¶
        #mask=color_mask,
        #å…è®¸æœ€å¤§è¯æ±‡
        max_words=5000,
        #æœ€å¤§å·å­—ä½“
        max_font_size=40
    ).generate_from_frequencies(text)#æ ¹æ®è¯é¢‘æ¥ç”»å›¾ï¼Œ https://blog.csdn.net/MIKASA3/article/details/80066001
    # { word1: fre1, word2: fre2,  word3: fre3,......,  wordn: fren }
    
    # matplotlibçš„æ–¹å¼å±•ç¤ºç”Ÿæˆçš„è¯äº‘å›¾åƒ
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
def collect_specified(comments, name):
    df = pd.DataFrame()
    for n in name:
        c = comments[comments['å†…å®¹'].str.contains(n)]
        df = pd.concat([df, c])
    df = df.drop_duplicates()
    return df
```

```python
def zhanbi(value, label):
    values = [1-sum(value),]
    values.extend(value)
    labels = ['å…¶ä»–',]
    labels.extend(label)
    x = range(1, len(values)+1)
    #ç”»æŸ±çŠ¶å›¾
    plt.rcParams['font.sans-serif'] = ['SimHei'] # æ­¥éª¤ä¸€ï¼ˆæ›¿æ¢sans-serifå­—ä½“ï¼‰
    plt.rcParams['axes.unicode_minus'] = False
    fig, ax = plt.subplots()
    for i in range(len(x)):
        b = ax.bar(x=x[i], height=values[i], width=0.4,
                   label=labels[i], color='darkorange')
        #h = b[0].get_height()
        h = values[i]
        ax.text(b[0].get_x()+b[0].get_width()/2, h,
                '%.2f%%'%(h*100), ha='center', va='bottom')
    plt.xticks(x, labels)
    plt.show()
```
